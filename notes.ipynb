```python
"""Sequence data types"""
#String
#A string is a sequence of characters enclosed in single quotes, double quotes, or triple quotes.
#Strings are immutable, meaning they cannot be changed after creation.
#They contain negative indexing as well like lists and tuples.
str1="Hello, World!"    
print(str1)
print(type(str1))
#String indexing
print(str1[0])  #H
print(str1[-1]) #!

#len() function to get length of string
print(len(str1))

#traversing string using for loop
for char in str1:
    print(char)
#traversing string using while loop
i=0
while i<len(str1):
    print(str1[i])
    i+=1
#String slicing and indexing
#syntax: string[start:end:step]
substr=str1[0:5]  #Hello
print(substr)   
substr=str1[7:]   #World!
print(substr)

#membership operator in and not in
print('H' in str1)  #True
print('h' in str1)  #False
print('x' not in str1)  #True

#String class
#view all methods of string class
print(dir(str))

#some commonly used methods
#find() method - returns the lowest index of substring if found else -1
#syntax: string.find(substring, start, end)
str2="Hello, welcome to the world of Python programming."
index=str2.find("world")
print(index)  #18
index=str2.find("Java")
print(index)  #-1 

#index() method - returns the lowest index of substring if found else raises ValueError
#syntax: string.index(substring, start, end)    
str2="Hello, welcome to the world of Python programming."
index=str2.index("Python")
print(index)  #31
#index=str2.index("Java")   
#print(index)  #raises ValueError

#rfiind() and rindex() methods - similar to find() and index() but search from right to left
str2="Hello, welcome to the world of Python programming. Python is great."
index=str2.rfind("Python")
print(index)  #49
index=str2.rindex("Python")
print(index)  #49

#alignment and padding methods - ljust(), rjust(), center(), zfill() - used to align and pad strings 
#these create new strings and return them
#syntax: string.ljust(width, fillchar), string.rjust(width, fillchar), string.center(width, fillchar), string.zfill(width)
str3="Hello"
print(str3.ljust(10,'*'))  #Hello*****
print(str3.rjust(10,'*'))  #*****Hello
print(str3.center(11,'*')) #***Hello****    
print(str3.zfill(10))      #00000Hello

#striping methods - lstrip(), rstrip(), strip() - remove leading and trailing characters
#these create new strings and return them
#syntax: string.lstrip(chars), string.rstrip(chars), string.strip(chars)
str4="***Hello World***"
print(str4.lstrip('*'))  #Hello World***
print(str4.rstrip('*'))  #***Hello World
print(str4.strip('*'))   #Hello World
#By default whitespace characters are removed if chars parameter is not provided
str5="   Hello World   "
print(str5.strip())  #Hello World
#You can also remove specific characters
str6="xyxxyHello Worldyyx"
print(str6.strip('xy'))  #Hello World


#join and split methods

#replace (old, new, count) - replaces occurrences of a substring with another substring
#syntax: string.replace(old, new, count)
str7="Hello World"
print(str7.replace("World", "Python"))  #Hello Python
print(str7.replace("l", "L"))  #HeLLo WorLd
print(str7.replace("l", "L", 1))  #HeLLo World

#join() method - joins elements of an iterable (list, tuple, etc.) into a single string
#syntax: separator.join(iterable)
str_list=["Hello","World","from","Python"]
joined_str=" ".join(str_list)
print(joined_str)  #Hello World from Python
print("-".join(str_list))  #Hello-World-from-Python

#split() method - splits a string into a list of substrings based on a separator
#syntax: string.split(separator, maxsplit)
str8="Hello World from Python"
substr_list=str8.split(" ")
print(substr_list)  #['Hello', 'World', 'from', 'Python']
substr_list=str8.split(" ",2)
print(substr_list)  #['Hello', 'World', 'from Python']

#other methods:

#startswith() and endswith() methods - check if string starts or ends with a substring
#syntax: string.startswith(substring, start, end), string.endswith(substring, start, end)
str9="Hello World"  
print(str9.startswith("Hello"))  #True
print(str9.endswith("World"))    #True
print(str9.startswith("World",6))  #True
print(str9.endswith("Hello",0,5))  #True

#removeprefix() and removesuffix() methods - remove a prefix or suffix from a string
#syntax: string.removeprefix(prefix), string.removesuffix(suffix)
str10="HelloWorld"
print(str10.removeprefix("Hello"))  #World
print(str10.removesuffix("World"))  #Hello

#partition() and rpartition() methods - split a string into three parts based on a separator
#syntax: string.partition(separator), string.rpartition(separator)
str11="Hello World from Python"
part1= str11.partition("World")
print(part1)  #('Hello ', 'World', ' from Python')
part2= str11.rpartition("from")
print(part2)  #('Hello World ', 'from', ' Python')


#capitalize() method - capitalizes the first character of the string
#upper() and lower() methods - convert string to uppercase or lowercase
#title() method - converts the first character of each word to uppercase
#swapcase() method - swaps the case of each character in the string
#casefold() method - converts string to lowercase for caseless comparisons. different from lower()
# in terms of handling special characters. e.g., German ß character is converted to 'ss' in casefold() but remains 'ß' in lower(). Thus applicable for all languages.
str12="hello world from python"
print(str12.capitalize())  #Hello world from python
print(str12.upper())       #HELLO WORLD FROM PYTHON
print(str12.lower())       #hello world from python
print(str12.title())       #Hello World From Python
print(str12.swapcase())    #HELLO WORLD FROM PYTHON
print(str12.casefold())    #hello world from python 


#inquality methods - isalpha(), isdigit(), isalnum(), isspace() - whitespace check, islower(), isupper(), istitle(), 
#isprintable()- return false if string contains non printable characters i.e white space, tabs, new line "\n\f\t" etc.
#isdecimal()-   checks if all characters in the string are decimal characters
# isascii()- checks if all characters in the string are ASCII characters
# isidentifier()- checks if the string is a valid identifier
#these methods return True if all characters in the string satisfy the condition else False
str13="HelloWorld"
str14="Hello123"
str15="   "
print(str13.isalpha())  #True
print(str14.isalnum())  #True
print(str14.isdigit())  #False
print(str15.isspace())  #True
print(str13.islower())  #False
print(str13.isupper())  #False
print(str13.istitle())  #True
print(str13.isprintable())  #True
print("123".isdecimal())  #True
print(str13.isascii())  #True
print("var_1".isidentifier())  #True 
```

```python
#Regular expressions

#string pattern matching and manipulation using re module
#generating patterns, searching, replacing, splitting strings based on patterns
#useed in data validation, parsing, text processing, web scraping, etc.

#1. pattern - 
# "and" - S1="android", S2="candy" - matches in both strings 'and'
#'com|org' - matches either 'com' or 'org' in a string. eg., S="example.com" - matches 'com'
#'[abcd]' - matches any character from the set {a, b, c, d} only one character. eg., S="cat" - matches 'c'
#'[abc]+'- matches one or more occurrences of characters from the set {a, b, c}. eg., S="aaabbbccc" - matches 'aaabbbccc'




#Quntifiers and metacharacters:
#'and' - matches the exact substring 'and'
#'.' - matches any single character except newline
#'|' - logical OR operator
#'^' - matches the start of the string
#'$' - matches the end of the string
#'\d' - matches any digit (0-9)
#'\D' - matches any non-digit
#'\w' - matches any alphanumeric character (a-z, A-Z, 0-9, _)
#'\W' - matches any non-alphanumeric character  
#'\s' - matches any whitespace character (space, tab, newline)
#'\S' - matches any non-whitespace character
#'*' - matches 0 or more occurrences of the preceding element
#'+' - matches 1 or more occurrences of the preceding element
#'?' - matches 0 or 1 occurrence of the preceding element
#'{n}' - matches exactly n occurrences of the preceding element
#'{n,}' - matches n or more occurrences of the preceding element
#'{n,m}' - matches between n and m occurrences of the preceding element
#'()' - groups multiple elements together
#[...] - matches any character inside the brackets
#[^...] - matches any character not inside the brackets 
#\A - matches the start of the string
#\Z - matches the end of the string 


#example patterns:
"""
'a+' - a, aa, aaa, aaaa, ...
'b*' - '', b, bb, bbb, bbbb, ...
'c?' - '', c
'd{3}' - ddd
'e{2,}' - ee, eee, eeee, ... 
'f{2,4}' - ff, fff, ffff #2 to 4 f's
'(abc)+' - abc, abcabc, abcabcabc, ... #same pattern 'abc' repeated one or more times
'[ab]' - a or b
'[a-z]' - any lowercase letter from a to z  
'[ab]+' - a, b, aa, ab, ba, bb, aaa, aab, aba, abb, baa, bab, bba, bbb, ... #one or more occurrences of a or b
'[0-9]{2,4}' - matches any number with 2 to 4 digits. eg., 12, 345, 6789
'[abc]?' - '', a, b, c
'[abc]*' - '', a, b, c, aa, ab, ac, ba, bb, bc, ca, cb, cc, aaa, aab, ... #zero or more occurrences of a,b,c
'(a|b)c' - ac or bc
'[a-f]' - any lowercase letter from a to f
'[^a-f]' - any character except lowercase letters from a to f   
'[a,b,c]' - a or b or c
'[a-zA-Z0-9]' - any alphanumeric character
'[a-z]+' - all lowercase words
'[A-Z]+' - all uppercase words
'[a-z][A-Z]' - two letter words with first letter lowercase and second letter uppercase 
[^a-z] - any character except lowercase letters
to include dot . in pattern \.

First name last name
Anmol Singh
'[A-Z][a-z]+ [A-Z][a-z]+'

Variable name: '[a-zA-Z_][a-zA-Z0-9]*'

time- 11:30 '[012][0-9]:[0-5][0=9]'

domain udemy.com- '[a-zA-Z0-9]+\.(com|org|net)$'
"""

#module re functions:
#compile(pattern) - compiles a regex pattern into a regex object
#Methods of regex object:
#what is match object ? 
#match object is an instance of the Match class returned by methods like search(), match(), fullmatch() etc.
#It contains information about the match, such as the matched string, start and end positions, etc.
#group() method of match object returns the matched string. 


#search(pattern, string) - searches for the pattern in the string and returns a match object
#match(pattern, string) - checks if the pattern matches the start of the string and returns a match object
#fullmatch(pattern, string) - checks if the pattern matches the entire string and returns a match object
#findall(pattern, string) - returns a list of all non-overlapping matches of the pattern in the string
#split(pattern, string) - splits the string at each match of the pattern and returns a list of substrings
#sub(pattern, repl, string) - replaces all occurrences of the pattern in the string with repl and returns the modified string
#example to explain above methods
import re
pattern=re.compile(r"and")
string1="android and candy"
match_obj=pattern.search(string1)
if match_obj:
    print("Pattern found using search():", match_obj.group())  #and 
match_obj=pattern.match(string1)
if match_obj:
    print("Pattern found using match():", match_obj.group())  #and
match_obj=pattern.fullmatch(string1)
if match_obj:
    print("Pattern found using fullmatch():", match_obj.group())
else:
    print("Pattern not found using fullmatch()")  #Pattern not found using fullmatch()
matches_list=pattern.findall(string1)
print("Matches found using findall():", matches_list)  
split_list=pattern.split(string1)   
print("String split using split():", split_list)  
modified_string=pattern.sub("XYZ", string1)
print("String after substitution using sub():", modified_string)    



```
