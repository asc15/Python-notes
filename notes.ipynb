```python
"""Data types in python""" 
"""a. Numeric data types - int, float, complex
b. Sequence data types - str, list, tuple 
c. Mapping data type - dict
d. Set data type - set, frozenset
e. Boolean data type - bool
f. Binary data types - bytes, bytearray, memoryview"""
#1. int
x=10
print(type(x))

#int can hold long values as well of any length
y=1234567890123456789012345678901234567890  
print(type(y))

#memory allocation of int is dynamic based on the value assigned
import sys
print(sys.getsizeof(x)) # size of int object in bytes
print(sys.getsizeof(y))

#int is immutable i.e, value cannot be changed. A new object is created in memory when value is changed
x=10    
print(id(x))
x=20
print(id(x))
#different memory locations  
#how to calculate int size in memory in bytes
import math
def int_size(n):
    if n==0:
        return 1
    else:
        return math.floor(math.log2(abs(n)))//8 +1
print(int_size(10))
print(int_size(12345678901234567890)) 
#difference in size calculated by sys.getsizeof and int_size is due to overhead of object in python. 
#overhead means extra memory used by python to manage the object. 
```

```python
#Base conversions
a=10  # decimal
b=0b1010 # binary       
c=0o12   # octal
d=0xA    # hexadecimal  
print(a,b,c,d)
#All will print 10  
#To convert from one base to another we can use int() function
num=25  # decimal
bin_num=bin(num)  # binary  
print(bin_num)
oct_num=oct(num)  # octal   
print(oct_num)
hex_num=hex(num)  # hexadecimal
print(hex_num)
#To convert from other bases to decimal we can use int() function with base parameter
bin_str='11001' # binary string
dec_from_bin=int(bin_str,2)
print(dec_from_bin)
oct_str='31'  # octal string
dec_from_oct=int(oct_str,8) 
print(dec_from_oct)
hex_str='1F'  # hexadecimal string
dec_from_hex=int(hex_str,16)
print(dec_from_hex)
#All will print 25
#To convert from one base to another directly we can use following approach
num=25  # decimal
bin_num=bin(num)  # binary
oct_num=oct(num)  # octal
hex_num=hex(num)  # hexadecimal
print(bin_num,oct_num,hex_num)
#To convert back to decimal
dec_from_bin=int(bin_num,2)
dec_from_oct=int(oct_num,8)
dec_from_hex=int(hex_num,16)
print(dec_from_bin,dec_from_oct,dec_from_hex)
```

```python
#Type conversion
#Implicit type conversion - done by python automatically
a=10    # int
b=12.5  # float
c=a+b  # int + float = float
print(c)
#Explicit type conversion - done by user using built-in functions
x=10    # int   
y=float(x)  # convert int to float
print(y)
# similarly we have int(), complex(), str(), list(), tuple(), dict(), set() functions for type conversion
#Notes: int(), float() It does not work with complex numbers but bool() works with all data types.
bool(2+3j) # True 
#Note: If we pass anything with bool it will give true, it only returns false for 0,0.0,"",[],(),{},set(),False
#Just like bool(), complex() also works with all data types except set and frozenset.
complex(10) # (10+0j)
complex(False) # 0j 
#Str() works with all data types.
str(-1.2E-3) # '-0.0012'
```

```python
"""Sequence data types"""
#String
#A string is a sequence of characters enclosed in single quotes, double quotes, or triple quotes.
#Strings are immutable, meaning they cannot be changed after creation.
#They contain negative indexing as well like lists and tuples.
str1="Hello, World!"    
print(str1)
print(type(str1))
#String indexing
print(str1[0])  #H
print(str1[-1]) #!

#len() function to get length of string
print(len(str1))

#traversing string using for loop
for char in str1:
    print(char)
#traversing string using while loop
i=0
while i<len(str1):
    print(str1[i])
    i+=1
#String slicing and indexing
#syntax: string[start:end:step]
substr=str1[0:5]  #Hello
print(substr)   
substr=str1[7:]   #World!
print(substr)

#membership operator in and not in
print('H' in str1)  #True
print('h' in str1)  #False
print('x' not in str1)  #True

#String class
#view all methods of string class
print(dir(str))

#some commonly used methods
#find() method - returns the lowest index of substring if found else -1
#syntax: string.find(substring, start, end)
str2="Hello, welcome to the world of Python programming."
index=str2.find("world")
print(index)  #18
index=str2.find("Java")
print(index)  #-1 

#index() method - returns the lowest index of substring if found else raises ValueError
#syntax: string.index(substring, start, end)    
str2="Hello, welcome to the world of Python programming."
index=str2.index("Python")
print(index)  #31
#index=str2.index("Java")   
#print(index)  #raises ValueError

#rfiind() and rindex() methods - similar to find() and index() but search from right to left
str2="Hello, welcome to the world of Python programming. Python is great."
index=str2.rfind("Python")
print(index)  #49
index=str2.rindex("Python")
print(index)  #49

#alignment and padding methods - ljust(), rjust(), center(), zfill() - used to align and pad strings 
#these create new strings and return them
#syntax: string.ljust(width, fillchar), string.rjust(width, fillchar), string.center(width, fillchar), string.zfill(width)
str3="Hello"
print(str3.ljust(10,'*'))  #Hello*****
print(str3.rjust(10,'*'))  #*****Hello
print(str3.center(11,'*')) #***Hello****    
print(str3.zfill(10))      #00000Hello

#striping methods - lstrip(), rstrip(), strip() - remove leading and trailing characters
#these create new strings and return them
#syntax: string.lstrip(chars), string.rstrip(chars), string.strip(chars)
str4="***Hello World***"
print(str4.lstrip('*'))  #Hello World***
print(str4.rstrip('*'))  #***Hello World
print(str4.strip('*'))   #Hello World
#By default whitespace characters are removed if chars parameter is not provided
str5="   Hello World   "
print(str5.strip())  #Hello World
#You can also remove specific characters
str6="xyxxyHello Worldyyx"
print(str6.strip('xy'))  #Hello World


#join and split methods
#join() method - joins elements of an iterable (list, tuple, etc.) into a single string
#syntax: separator.join(iterable)
```
