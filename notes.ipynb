```python
"""Data types in python""" 
"""a. Numeric data types - int, float, complex
b. Sequence data types - str, list, tuple 
c. Mapping data type - dict
d. Set data type - set, frozenset
e. Boolean data type - bool
f. Binary data types - bytes, bytearray, memoryview"""
#1. int
x=10
print(type(x))

#int can hold long values as well of any length
y=1234567890123456789012345678901234567890  
print(type(y))

#memory allocation of int is dynamic based on the value assigned
import sys
print(sys.getsizeof(x)) # size of int object in bytes
print(sys.getsizeof(y))

#int is immutable i.e, value cannot be changed. A new object is created in memory when value is changed
x=10    
print(id(x))
x=20
print(id(x))
#different memory locations  
#how to calculate int size in memory in bytes
import math
def int_size(n):
    if n==0:
        return 1
    else:
        return math.floor(math.log2(abs(n)))//8 +1
print(int_size(10))
print(int_size(12345678901234567890)) 
#difference in size calculated by sys.getsizeof and int_size is due to overhead of object in python. 
#overhead means extra memory used by python to manage the object. 
```

```python
#2. float
f=12.5
print(type(f))
print(sys.getsizeof(f)) # size of float object in bytes 
#size of float is fixed 
#float is immutable
f=12.5
print(id(f))
f=15.5
print(id(f))
#different memory locations 
#how to calculate size of float
import struct
print(struct.calcsize('d')) # size of double precision float in bytes 

a=125E2  # scientific notation (mantissa E exponent)
print(a)
a=0.25e-2
print(a) 
```

```python
#bool
b=True
print(type(b))
print(sys.getsizeof(b)) # size of bool object in bytes
#Integers are used to represent boolean values internally in python
print(int(True))  # 1
```
